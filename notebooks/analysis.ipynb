```python
# Анализ артистов
# Посмотрим самых популярных исполнителей по популярности и количеству подписчиков
top_artists = df.groupby('artist_name')[['artist_popularity', 'artist_followers']].max().sort_values('artist_popularity', ascending=False).head(10)
top_artists
# Наблюдения:
# Лидеры по популярности и по подписчикам не всегда совпадают
# Значит, популярность трека и база подписчиков - связанные, но не одинаковые параметры

```

```python
# Топ 10 песен по популярности
top_tracks = df.sort_values('track_popularity', ascending=False).head(10)
top_tracks[['track_name','artist_name', 'track_popularity']]
```

```python
# Как менялась популярность с годами?
df['track_popularity'] = pd.to_numeric(df['track_popularity'], errors='coerce')
df['year'] = pd.to_datetime(df['album_release_date'], errors='coerce').dt.year
df.groupby('year')['track_popularity'].mean().plot(figsize=(9,4))
plt.title('Динамика популярности музыки по годам')
plt.ylabel('Средняя популярность')
# Ранние годы (до 1980-х) показывают постепенный рост популярности
# 1980-е выделяются как период пиковых значений
# Современная музыка (2000–2024) относительно стабильна по популярности
```

```python
# топ 10 артистов по подписчикам
top_artists = df.groupby('artist_name').agg({'artist_followers':'max','artist_popularity':'max'}).sort_values('artist_followers', ascending=False).head(10).reset_index()
top_artists[['artist_name','artist_followers', 'artist_popularity']]
```

```python
# Статистика по длительности
duration_info = df[['track_duration_min']].describe().round(2)
duration_info
```

```python
# Самые короткие треки
df['track_duration_min'] = pd.to_numeric(df['track_duration_min'], errors='coerce')
shortest = df[(df['track_duration_min']< 2)].sort_values('track_duration_min').head(10)
shortest[['track_name', 'artist_name', 'track_duration_min']]
```

```python
# Самые длинные треки
longest = df[(df['track_duration_min']>5)].sort_values('track_duration_min',  ascending=False).head(10)
longest[['track_name', 'artist_name', 'track_duration_min']]
```

```python
# Зависит ли популярность песен от длительности?
df['duration_group'] = pd.cut(df['track_duration_min'],
                              bins=[0,2.5,4,6,20],
                              labels=['короткие', 'средние', 'длинные', 'очень длинные'])

df.groupby('duration_group')['track_popularity'].mean().sort_values(ascending=False).reset_index()
# Популярность слабо зависит от длительности. Однако у популярных песен чаще стандартная длина 2,5–4 мин - это формат, удобный для
# стриминга и радио
# Слишком короткие и длинные треки становятся хитами реже

```

```python
# Визуализация зависимости
plt.figure(figsize=(10,10))
plt.scatter(df['track_duration_min'], df['track_popularity'], alpha=0.3)
plt.title('Связь длительности трека и популярности')
plt.xlabel('Длительность (мин)')
plt.ylabel('Популярность')
```

```python
# Топ 10 самых популярных жанров
df_genres = df.copy()
df_genres['artist_genres'] = df_genres['artist_genres'].fillna('Unknown')
df_genres['artist_genres'] = df_genres['artist_genres'].apply(lambda x: x.split(', '))
df_genres = df_genres.explode('artist_genres')
top_genres = df_genres.groupby('artist_genres')['track_popularity'].mean().sort_values(ascending=False).head(10).reset_index()
top_genres
# Мы сгруппировали данные по жанрам и вычислили их среднюю популярность
# Перед анализом заполнили пропуски значением “Unknown” и преобразовали колонку
# artist_genres в список жанров с помощью explode()

```

```python
# Зависит ли популярность от типа альбома ?
df.groupby('album_type')['track_popularity'].mean().reset_index()
# Треки из альбомов более популярны, чем синглы и сборники
```
